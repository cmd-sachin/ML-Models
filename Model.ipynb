```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
feature = pd.read_csv("X_Train_Data_Input.csv")
target = pd.read_csv("Y_Train_Data_Target.csv")

test_feature = pd.read_csv("X_Test_Data_Input.csv")
test_target = pd.read_csv("Y_Test_Data_Target.csv")
```

```python
na_info = feature.isna().sum()
feature.fillna(0,inplace=True)
feature.to_csv("X_Train_Data_Input.csv",index = False)
na_info = test_feature.isna().sum()
test_feature.fillna(0,inplace=True)
test_feature.to_csv("X_Test_Data_Input.csv",index=False)
```

```python
X = np.array([feature['Column0'],feature['Column1'],feature['Column2'],feature['Column3'],feature['Column4'],feature['Column5'],feature['Column6'],feature['Column7'],feature['Column8'],feature['Column9'],feature['Column10'],feature['Column11'],feature['Column12'],feature['Column13'],feature['Column14'],feature['Column15'],feature['Column16'],feature['Column17'],feature['Column18'],feature['Column19'],feature['Column20'],feature['Column21']]).reshape(-1,22)
y = np.array(target['target']).reshape(-1,)

X_test = np.array([test_feature['Column0'],test_feature['Column1'],test_feature['Column2'],test_feature['Column3'],test_feature['Column4'],test_feature['Column5'],test_feature['Column6'],test_feature['Column7'],test_feature['Column8'],test_feature['Column9'],test_feature['Column10'],test_feature['Column11'],test_feature['Column12'],test_feature['Column13'],test_feature['Column14'],test_feature['Column15'],test_feature['Column16'],test_feature['Column17'],test_feature['Column18'],test_feature['Column19'],test_feature['Column20'],test_feature['Column21']]).reshape(-1,22)
y_test = np.array(test_target['target']).reshape(-1,)
```

```python
scaler = preprocessing.MinMaxScaler()
X_norm = scaler.fit_transform(X)
X_test = scaler.fit_transform(X_test)
```

```python
model = KNeighborsClassifier(n_neighbors=17)
model.fit(X,y)
```

```python
prediction = model.predict(X_test)
```

```python
print(accuracy_score(y_test,prediction))
```
